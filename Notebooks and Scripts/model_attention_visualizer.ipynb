```python
from transformers import TFAutoModel, AutoTokenizer, AutoModelForSequenceClassification, DistilBertForSequenceClassification
tokenizer = AutoTokenizer.from_pretrained("./recent_twitter_model")
model = DistilBertForSequenceClassification.from_pretrained("./recent_twitter_model", output_attentions=True)
```

```python
#Will use a recent tweet from Ted Cruz as an example
inputs = tokenizer.encode("COVID positivity rates rising in Texas is a direct result of illegal aliens being released into communities. #BidenBorderCrisis", return_tensors='pt')
outputs = model(inputs)
attention = outputs[-1]  # Output includes attention weights when output_attentions=True
tokens = tokenizer.convert_ids_to_tokens(inputs[0]) 
```

```python
from bertviz import head_view
head_view(attention, tokens)
```
